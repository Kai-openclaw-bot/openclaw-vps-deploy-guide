# Развёртывание OpenClaw на VPS

## Полное руководство по self-hosting AI-ассистента


Приветствие. Сегодня разберём, как развернуть OpenClaw на собственном VPS — чтобы ассистент был доступен 24/7 через Telegram, WhatsApp и другие каналы.

## Что такое OpenClaw?

- - Self-hosted gateway для AI-агентов — работает на вашем железе
- - Multi-channel: Telegram, WhatsApp, Discord, Signal, Slack, iMessage и др.
- - Agent-native: сессии, память, multi-agent routing, skills
- - Open source (MIT), данные остаются у вас
- - Heartbeat: проактивный мониторинг, cron-задачи, автоматизация

OpenClaw — это не просто обёртка над LLM. Это полноценный gateway, который связывает мессенджеры с AI-агентом. Ключевое отличие от hosted-решений: полный контроль над данными.

## Архитектура: как это работает

- - Gateway — единый процесс, источник истины (state + workspace)
- - Channel plugins — подключение мессенджеров
- - Agent runtime — сессии, инструменты, память
- - Control UI — браузерный дашборд (порт 18789)
- - Nodes — iOS/Android для camera, canvas, voice

```python
# Visualization placeholder
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Архитектура: как это работает')
ax.text(0.5, 0.5, 'Add your chart here', ha='center', va='center', fontsize=16)
ax.set_xlim(0, 1); ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
```

Gateway — сердце системы. Один процесс обслуживает все каналы. State хранится в ~/.openclaw/. Nodes — опциональны, для мобильных возможностей.

## Требования к серверу

- - ОС: Linux (Ubuntu 22.04/24.04 рекомендуется)
- - RAM: минимум 2 ГБ, рекомендуется 4 ГБ
- - CPU: 2 ядра минимум
- - Диск: 20 ГБ SSD
- - Node.js: 22.19+ или 24+ (рекомендуется 24)
- - API-ключ LLM-провайдера (Anthropic, OpenAI, Google и др.)
- - SSH-доступ к серверу

```python
# Visualization placeholder
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Требования к серверу')
ax.text(0.5, 0.5, 'Add your chart here', ha='center', va='center', fontsize=16)
ax.set_xlim(0, 1); ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
```

Для запуска с локальными моделями через Ollama нужно больше RAM — от 8 ГБ. Для cloud API (Claude, GPT) 2 ГБ достаточно.

## Шаг 4: Безопасность и доступ

- - Gateway на loopback (127.0.0.1:18789) — не торчим наружу
- - SSH-туннель для дашборда:
-   ssh -N -L 18789:127.0.0.1:18789 openclaw@VPS_IP
- - Tailscale — zero exposed ports, encrypted access
- - Firewall: ufw allow 22, deny 18789
- - gateway.auth.token — обязательный shared secret

```python
# Visualization placeholder
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Шаг 4: Безопасность и доступ')
ax.text(0.5, 0.5, 'Add your chart here', ha='center', va='center', fontsize=16)
ax.set_xlim(0, 1); ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
```

Никогда не выставляйте порт 18789 в публичный интернет без auth-токена! Tailscale — лучший вариант: ни одного открытого порта, end-to-end шифрование.

## Тюнинг для слабых VPS

- export OPENCLAW_NO_RESPAWN=1
- 
# systemd tuning:
# Restart=always, RestartSec=2, TimeoutStartSec=90
- 
# SSD для state/cache путей
# Мониторинг OOM: dmesg | grep -i oom

```python
# Visualization placeholder
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Тюнинг для слабых VPS')
ax.text(0.5, 0.5, 'Add your chart here', ha='center', va='center', fontsize=16)
ax.set_xlim(0, 1); ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
```

На 2 ГБ RAM могут быть OOM-киллы. NODE_COMPILE_CACHE даёт 2-3x ускорение повторных запусков. OPENCLAW_NO_RESPAWN=1 избегает лишних process handoff.
